```
# LSTM

# Adding relevant libraries

import math 
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.utils import Sequence
from datetime import timedelta
from sklearn.metrics import mean_squared_error
import time
```

```
df_DEBJE.info()
```

```
# Splitting data into train and test set
df_DEBJE_model = df_DEBJE_timeseries.resample('M').sum().reset_index()
test_cutoff_date = df_DEBJE_model['date'].max() - timedelta(days = 450)

df_test = df_DEBJE_model[df_DEBJE_model['date'] > test_cutoff_date]
df_train = df_DEBJE_model[df_DEBJE_model['date'] <= test_cutoff_date]
df_test = df_test.reset_index()
df_test.drop(columns=['index'], inplace=True)
print('Test dates: {} to {}'.format(df_test['date'].min(), df_test['date'].max()))
print('Train dates: {} to {}'.format(df_train['date'].min(), df_train['date'].max()))
```

```
def create_ts_df(dataset, start_index, end_index, history_length, step_size, target_step):
  assert step_size > 0
  assert start_index >= 0
  
  time_lags = sorted(range(target_step+1, target_step+history_length+1, step_size), reverse=True)
  col_names = [f'x_lag{i}' for i in time_lags] + ['y']
  start_index = start_index + history_length
  
  if end_index is None:
    end_index = len(dataset) - target_step
    
  data_list = []
  
  ind0 = history_length
  ind1 = len(dataset) - 1
  
  for j in range(ind0, ind1):
    indices = range(j-1, j-history_length-1, -step_size)
    data = dataset[sorted(indices) + [j]]
    data_list.append(data)
    
  df_ts = pd.DataFrame(data=data_list, columns=col_names)

  return len(col_names) - 1, df_ts
```

```
num_timesteps, df_all = create_ts_df(df_DEBJE_model['orders'].values, start_index=0, end_index=None, history_length=12, step_size=1, target_step=1)
df_all
```

```
num_timesteps, df_ts = create_ts_df(df_train['orders'].values, start_index=0, end_index=None, history_length=12, step_size=1, target_step=1)
```

```
df_ts
```

```
ts_inputs = tf.keras.Input(shape=(num_timesteps, 1))
x = layers.LSTM(units=20)(ts_inputs)
x = layers.Dropout(0.2)(x)
outputs = layers.Dense(1, activation='linear')(x)
model = tf.keras.Model(inputs=ts_inputs, outputs=outputs)
```

```
# training configuration - optimization function and loss function

model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.01), 
             loss=tf.keras.losses.MeanSquaredError(),
             metrics=['mse'])
```

```
model.summary()
```

```
import numpy as np
features = df_ts.drop('y', axis=1).values
target = df_ts['y'].values
num_records = len(df_ts.index)
features_train = np.array(features).reshape(num_records, -1, 1)
model.fit(x=features_train, y=target, epochs=100)
```

```
num_timesteps, df_test_ts = create_ts_df(df_DEBJE_model['orders'].values, start_index=0, end_index=None, history_length=12, step_size=1, target_step=1)
features = df_test_ts.drop('y', axis=1).values
features_arr = np.array(features)
num_records = len(df_test_ts.index)
features_test = features_arr.reshape(num_records, -1, 1)
y_pred = model.predict(features_test).reshape(-1, )

y_act = df_test_ts['y'].values
print('test mean squared error: {}'.format(mean_squared_error(y_act, y_pred)))

#baseline
y_pred_baseline = df_test_ts['x_lag11'].values
print('test baseline mean squared error: {}'.format(mean_squared_error(y_act, y_pred_baseline)))
```

```
from sklearn.metrics import mean_absolute_error
print('test mean absolute error: {}'.format(mean_absolute_error(y_act, y_pred)))
```
